# Sensordiagnose im Steuergerät

Zur Erkennung von Sensorfehlern werden in Steuergeräten Diagnoseverfahren eingesetzt. Sie überwachen Ausgangsspannungen von Sensoren und plausibilisieren unterschiedliche Sensorsignale miteinander. Es wird die Frage diskutiert, mit welcher Wahrscheinlichkeit ein Sensor wirklich defekt ist, wenn er mit der oben diskutierten Diagnosefunktion als defekt erkannt wurde. Um die Aufgabe zu lösen, werden zu Beginn zwei Ereignisse definiert:

Ereignis A: Sensor ist defekt

Ereignis B: Diagnoseergebnis ist positiv, Sensor wird als defekt eingestuft

Die Wahrscheinlichkeit für einen defekten Sensor bei einem positiven Diagnoseergebnis kann nach dem Satz von Bayes über folgende Gleichung berechnet werden. 

$P\left( {A|B} \right) = \frac{{P\left( {B|A} \right) \cdot P\left( A \right)}}{{P\left( B \right)}}$

Der Anteil funktionsfähiger Sensoren ist typischerweise sehr groß ist. Deshalb ist die bedingte Wahrscheinlichkeit für einen angezeigten Defekt bei funktionsfähigem Sensor sehr wichtig. 

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import matplotlib
SMALL_SIZE = 14
matplotlib.rc('font', size=SMALL_SIZE)
matplotlib.rc('axes', titlesize=SMALL_SIZE)

# Generate dataset
PD = 1e-4
PND = 1-1e-4
PED_D = 0.9999
PED_ND = 0.0002

x = np.logspace(-4, 0, 50)
P_D_ED = PED_D*PD / (PED_D*PD+x/100*PND)

# Plot results
fig, ax = plt.subplots(figsize=(7, 5))
ax.semilogx(x, P_D_ED, 'C0')
ax.semilogx(0.02, 0.333, 'C0o', markersize=6)
ax.grid(True)
ax.axis([1e-4, 1, 0, 1])
ax.set_xlabel("Wahrscheinlichkeit P(B|A') / %")
ax.set_ylabel('Aussagesicherheit P(A|B) / %')
```

Die Aussagesicherheit steigt mit fallender Wahrscheinlichkeit für einen angezeigten Defekt bei funktionsfähigem Sensor.
